```python
train = pd.read_csv('training.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
train.columns     # 칼럼명 확인
```

# 2. train데이터와 test 데이터 합치기

```python
alldata = pd.concat([train, test])
alldata
```

# 3. 필요없는 칼럼 삭제(순번 칼럼, 정답 칼럼, 날짜 칼럼 지우기)

```python
alldata2 = alldata.drop(columns = ['RefId', 'IsBadBuy', 'PurchDate'])
alldata2
```

# 4. 문자형 칼럼의 변수명 확인(unique)하고 숫자로 대체(replace)

```python
alldata2['Auction'].unique()
```

```python
alldata2['Auction'] = alldata2['Auction'].replace({'ADESA' : 1, 'OTHER' : 2, 'MANHEIM' : 3})
alldata2
```

# 5. 문자형 칼럼의 변수명이 많을 때 숫자로 대체하는 방법: LabelEncoder 사용
- 변수명 확인: unique()

```python
alldata2['Make'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
alldata2['Make'] = le.fit_transform(alldata2['Make'])
alldata2
```

```python
# 칼럼 전체 다 보이게

pd.options.display.max_columns = 999
alldata2
```
