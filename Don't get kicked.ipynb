```python
train = pd.read_csv('training.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
train.columns     # 칼럼명 확인
```

# 2. train데이터와 test 데이터 합치기

```python
alldata = pd.concat([train, test])
alldata
```

# 3. 필요없는 칼럼 삭제(순번 칼럼, 정답 칼럼, 날짜 칼럼 지우기)

```python
alldata2 = alldata.drop(columns = ['RefId', 'IsBadBuy', 'PurchDate'])
alldata2
```

# 4. 문자형 칼럼의 변수명 확인(unique)하고 숫자로 대체(replace)

```python
alldata2['Auction'].unique()
```

```python
alldata2['Auction'] = alldata2['Auction'].replace({'ADESA' : 1, 'OTHER' : 2, 'MANHEIM' : 3})
alldata2
```

# 5. 문자형 칼럼의 변수명이 많을 때 숫자로 대체하는 방법: LabelEncoder 사용
- 변수명 확인: unique()

```python
alldata2['Make'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
alldata2['Make'] = le.fit_transform(alldata2['Make'])
alldata2
```

```python
# 칼럼 전체 다 보이게

pd.options.display.max_columns = 999
alldata2
```

# 6. 문자형 칼럼만 확인

```python
c = alldata2.columns[alldata2.dtypes == 'object']
c
```

# 7. 모든 문자형 칼럼을 숫자로 변환

```python
alldata2.isnull().sum()     # 결측치 확인
```

```python
for i in c:
  alldata2[i] = le.fit_transform(list(alldata2[i]))     # 코랩에서는 list()를 안해도 에러나지 않지만 캐글에서는 에러발생
                                                        # NaN값이 float로 되기 때문 
alldata2
```

```python
alldata2.isnull().sum()     # 결측치 다시 확인
```

```python
# 결측치 채우기

alldata2 = alldata2.fillna(-1)     # 기존 데이터의 0 또는 평균값과 동일한 값과 구분하기 위해 결측치를 0이나 평균값이 아닌 -1로 대체
```

```python
alldata2.isnull().sum()
```

# 8. train2, test2로 분리

```python
train2 = alldata2[:len(train)]     # alldata2를 앞에서부터 train 갯수만큼 train2로 분리
test2 = alldata2[len(train):]     # alldata2를 train갯수 이후부터 나머지를 test2로 분리
```

# 9. 모델링(모델 불러오기, 모델 선언, 모델 학습, 결과 예측)

```python
from sklearn.ensemble import RandomForestClassifier     # 분류문제이므로 RandomForestClassifier 사용

rf = RandomForestClassifier(n_jobs = -1, random_state = 42)     # n_jobs = -1은 cpu를 최대로 사용, random_state 지정 안하면 할 때마다 랜덤하게 변하므로 일정하게 하기 위해 아무 숫자나 지정
rf.fit(train2, train['IsBadBuy'])     # 정답칼럼: IsBadBuy
result = rf.predict_proba(test2)     # 캐글 분류대회는 확률로 제출해야 하기 때문에 결과 예측을 predict_proba로 함
result     # result 결과 중 오른쪽(클래스1 = True)에 있는 걸 제출해야 함(첫 줄에서 0.95, 0.05 중 0.05)
```

- cf) result에서 0과 1의 갯수를 세려면 판다스로 바꾼 후 갯수 세야함
- pd.Series(result).value_counts()
- 캐글 분류대회에서는 확률로 제출할 것이므로 이 과정 필요 없음

# 10. 제출 양식 불러오기

# 11. 제출양식에 결과 포함시키기
